```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn import  datasets
import csv



C1_DATA = np.loadtxt("/home/rajorshi/ML/data/ref_stream_0.txt", comments="#", delimiter=" ", unpack=False)
C2_DATA = np.loadtxt("/home/rajorshi/ML/data/ref_stream_1.txt", comments="#", delimiter=" ", unpack=False)

C1_X1=C1_DATA[:,0];
C1_X2=C1_DATA[:,1];
C2_X1=C2_DATA[:,0];
C2_X2=C2_DATA[:,1];



accuracy=[];

start=5
end=20
for bs in range(start,end):

    BIN_SIZE=10;

    DATA=[];

    Y=[];
    for i in range(0, C1_DATA.shape[0]-BIN_SIZE):
        X1=C1_X1[i:i+BIN_SIZE];
        X2=C1_X2[i:i+BIN_SIZE];
        X=np.append(X1,X2);
        DATA.append(X);
        Y.append(0);


    for i in range(0, C2_DATA.shape[0]-BIN_SIZE):
        X1=C2_X1[i:i+BIN_SIZE];
        X2=C2_X2[i:i+BIN_SIZE];
        X=np.append(X1,X2);
        DATA.append(X);
        Y.append(1);

    DATA=np.array(DATA)



    X_train, X_test, Y_train, Y_test =  train_test_split(DATA, Y, test_size=.4, random_state=42)
    #print(np.array(X_train).shape, np.array(Y_train).shape)
    #print(np.array(X_test).shape, np.array(Y_test).shape)

    names = [
             "Nearest Neighbors",
             "Linear SVM", 
             "RBF SVM", 
             "Decision Tree", 
             "Random Forest", 
             "Neural Net", 
             "AdaBoost"
    ]
    names = [
        #     "k-NN",
        #     "L-SVM", 
         #    "RBF-SVM", 
          #   "DT", 
           #  "RF", 
             "NN", 
             #"AB"
    ]


    classifiers = [
      #  KNeighborsClassifier(bs),
      #  SVC(kernel="linear", C=0.025),
       # SVC(gamma=4, C=2),
        #DecisionTreeClassifier(max_depth=8),
        #RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        MLPClassifier(alpha=1,hidden_layer_sizes=bs, activation="relu"),
        #AdaBoostClassifier()
    ]



    values=[];
    for name, clf in zip(names, classifiers):
        clf.fit(X_train, Y_train)
        score = clf.score(X_test,Y_test)
        #print(name, score);
        predicted = clf.predict(X_test)
        given=Y_test; 
        values.append(score);


        fn=0;
        fp=0;
        tp=0
        tn=0
        for i in range(X_test.shape[0]):
            if predicted[i]==1  and given[i]==0:
                fp=fp+1;
            if predicted[i]==0  and given[i]==1:
                fn=fn+1;

            if predicted[i]==0  and given[i]==0:
                tp=tp+1;

            if predicted[i]==1  and given[i]==1:
                tn=tn+1;


        #print("false negative", fn/len(given))
        #print("false positive",fp/len(given))
        #print("true positive",tp/len(given))
        #print("true negative",tn/len(given))

        #objects=['FP', 'FN'];
        #performance=[fp/len(given),fn/len(given)];


       # plt.savefig("fp_fn_"+name+".png");

       # plt.show()
        
        
    
    accuracy.append(values);
    #print(values);

i=0
acc=np.array(accuracy).transpose();
for a in acc:
    plt.plot(range(start,end), a,label=names[i]);
    i=i+1;
print(acc)
plt.legend()
plt.savefig("nn-k.png")
plt.show();
```
